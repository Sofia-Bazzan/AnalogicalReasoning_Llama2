```python
# Response Variable for the accuracy

response_test = [] # List of responses
word_to_check = ["Cross-Domain", "Within-Domain", "FALSE"]

for item in test_data_human:
    for word in word_to_check:
        if word in item["inputs"]:
              response_test.append(word) # We append the reesponse to the list

print(response_test)
```

```python
from gradientai import Gradient

llama_response_before = [] # List of llama 2 before fine-tuning
llama_response_after = [] # List of Llama 2 afetr fine-tuning
count = 0

for index in range(len(test_data)):
    
    count = count + 1

    # adjust the string for the output of the completion
    # words_to_remove = ["The analogy drawn is between elements from different domains or categories", "The analogy drawn is within the same domain or category", "The analogy drawn is illogical or does not follow a valid relationship"]
    words_to_remove = ["Cross-Domain", "Within-Domain", "FALSE"]
    for word in words_to_remove:
        test_data[index]["inputs"] = test_data[index]["inputs"].replace(word, '')

    word_to_check = ["Cross-Domain", "Within-Domain", "FALSE"] # Words to check in order to save the reponse in the llama_response list

    word_to_add = "? Response with one among these: Cross-Domain or Within-Domain or FALSE" # We want that the Llama 2 model response with Cross-Domain, Within-Domain or FALSE
    test_data[index]["inputs"] = test_data[index]["inputs"].replace("?", word_to_add)

    with Gradient() as gradient:
        base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

        new_model_adapter = base_model.create_model_adapter(name = "test model") # Creation of the model
        print(f"Created model adapter with id: {new_model_adapter.id}") # Print of the model
        sample_query = test_data[index]["inputs"] # Prompt input
        print(f"{count}) Asking: {sample_query}")
        # before fine-tuning
        completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output
        print(f"Response Generated (before fine-tune): {completion}") # Print of the response before fine-tuning

        # Store the Response before fine-tuning
        for word in word_to_check:
            if word in completion:
                llama_response_before.append(word)

        samples = train_data

        # this is where fine-tuning happens
        # num_epochs is the number of times you fine-tune the model
        # more epochs tends to get better results, but you also run the risk of "overfitting"
        # play around with this number to find what works best for you
        num_epochs = 2
        i = 0
        while i < num_epochs:
            print(f"Fine-tuning the model, iteration {i + 1}")
            new_model_adapter.fine_tune(samples=samples)
            i = i + 1

        # after fine-tuning
        completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=100).generated_output # Generation of the Response after fine-tuning
        print(f"Response Generated (after fine-tune): {completion}") # print of the response after fine-tuning

        # Store the Response after fine-tuning
        for word in word_to_check:
            if word in completion:
                llama_response_after.append(word)

        print("---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")


        new_model_adapter.delete() # Deletion of the model for the next input in the next iteration

```

```python
# Calculate accuracy
correct_predictions = sum(a == p for a, p in zip(response_test, llama_response_before))
total_predictions = len(response_test)
accuracy = correct_predictions / total_predictions

print(f"Accuracy before Fine-Tuning: {accuracy * 100:.2f}%")

correct_predictions = sum(a == p for a, p in zip(response_test, llama_response_after))
total_predictions = len(response_test)
accuracy = correct_predictions / total_predictions

print(f"Accuracy after Fine-Tuning: {accuracy * 100:.2f}%")
```

```python
from gradientai import Gradient

llama_response_2 = [] # List of Llama 2 responses
count = 0

for index in range(len(data_for_llama)):

    count = count + 1

    # adjust the string for the output of the completion
    words_to_remove = ["The analogy drawn is between elements from different domains or categories", "The analogy drawn is within the same domain or category", "The analogy drawn is illogical or does not follow a valid relationship"]
    for word in words_to_remove:
        data_for_llama[index]["inputs"] = data_for_llama[index]["inputs"].replace(word, '')

    word_to_check = ["Cross-Domain", "Within-Domain", "FALSE", "cross-domain", "within-domain"] # Words to check in order to save the reponse in the llama_response list

    word_to_add = "? Choose one among these: Cross-Domain or Within-Domain or FALSE" # We want that the Llama 2 model response with Cross-Domain, Within-Domain or FALSE
    data_for_llama[index]["inputs"] = data_for_llama[index]["inputs"].replace("?", word_to_add)

    with Gradient() as gradient:
        base_model = gradient.get_base_model(base_model_slug="nous-hermes2")

        new_model_adapter = base_model.create_model_adapter(name = "test model") # Creation of the model
        print(f"Created model adapter with id: {new_model_adapter.id}") # Print of the model adapter
        sample_query = data_for_llama[index]["inputs"] # Prompt input
        print(f"{count}) Asking: {sample_query}")

        completion = new_model_adapter.complete(query=sample_query, max_generated_token_count=50).generated_output # Generation of the response
        print(f"Response Generated by the model: {completion}") # Print of the response

        # Store the Response
        for word in word_to_check:
            if word in completion:
                llama_response_2.append(word)
                break

        if len(llama_response_2) != count:
            word = random.choice(["Cross-Domain", "Within-Domain", "FALSE"])
            llama_response_2.append(word)

        print(len(llama_response_2))
        print("---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

        new_model_adapter.delete() # Deletion of the model for the next input in the next iteration
```

```python
# Adjust the list
for index in range(len(llama_response_2)):
    if llama_response_2[index] == "cross-domain":
        llama_response_2[index] = "Cross-Domain"
    if llama_response_2[index] == "within-domain":
        llama_response_2[index] = "Within-Domain"
```

```python
# Calculate accuracy
print(len(llama_response_2))
print(len(response))

correct_predictions = sum(a == p for a, p in zip(response, llama_response_2))
false_positive_within=0
false_positive_cross=0
false_negative=0
for i in range(len(llama_response_2)):
    if llama_response_2[i]=="Within-Domain" and (response[i]=="Cross-Domain" or response[i]=="FALSE"):
        false_positive_within+=1
    if llama_response_2[i]=="Cross-Domain" and (response[i]=="Within-Domain" or response[i]=="FALSE"):
        false_positive_cross+=1
    if  llama_response_2[i]=="FALSE" and (response[i]=="Within-Domain" or response[i]=="Cross-Domain"):
        false_negative+=1

total_predictions = len(response)
accuracy = correct_predictions / total_predictions
print(false_positive_within)
print(false_positive_cross)
print(false_negative)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
